```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
PI = math.pi
def sin(x, n):
    a = x
    for k in range(1, n):
        if k % 2 != 0:
            a -= (x**(2*k+1))/math.factorial((2*k+1))
        else:
            a += (x**(2*k+1))/math.factorial((2*k+1))
    return a

# print(PI)
# print(sin(PI/2, 20))
```

```python
def e(x, n):
    a = 1
    for k in range(1, n):
        a += (x**k)/math.factorial(k)
    return a
# print(e(1, 20))
```

```python
x = np.arange(0., 7, 0.1)
# print(x)
sin_re = np.sin(x)
for N in range(0, 25, 5):
    y_sin_T = np.array([sin(i, N) for i in x])
    plt.plot(x, y_sin_T)
    plt.title('sin_N='+str(N))
    plt.savefig('sin_N='+str(N)+'.png')
    plt.show()

    sin_error = np.fabs(sin_re - y_sin_T)
    plt.plot(x, sin_error)
    plt.title('sin_error_N='+str(N))
    plt.savefig('sin_error_N='+str(N)+'.png')
    plt.show()
```

```python
e_re = np.exp(x)
for N in range(0, 25, 5):
    y_e_T = np.array([e(i, N) for i in x])
    plt.plot(x, y_e_T)
    plt.title('e_N='+str(N))
    plt.savefig('e_N='+str(N)+'.png')
    plt.show()
    
    e_error = np.fabs(e_re - y_e_T)
    plt.plot(x, e_error)
    plt.title('e_error_N='+str(N))
    plt.savefig('e_error_N='+str(N)+'.png')
    plt.show()
```
