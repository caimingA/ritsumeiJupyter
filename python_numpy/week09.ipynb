```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# 課題1

```python
def f(x):
    return x**3 - (3/2) * x**2 - (3/2) * x + 1


def f_D(x):
    return 3 * x**2 - 3 * x - 3/2

x_n = 0
x_n_1 = 1
counter = 1
while math.fabs(x_n_1 - x_n) > 1e-6:
    x_n = x_n_1
    x_n_1 = x_n - (f(x_n))/(f_D(x_n))
    print("x",counter," = ", x_n_1)
    counter += 1
```

# 課題2

```python
def g(x):
    return math.exp(-x) +  x**2 - 6*x


def g_D(x):
    return -math.exp(-x) + 2 * x - 6

x_n = 0
x_n_1 = 1
counter = 1
while math.fabs(x_n_1 - x_n) > 1e-6:
    x_n = x_n_1
    x_n_1 = x_n - (g(x_n))/(g_D(x_n))
#     print("x",counter," = ", x_n_1)
    counter += 1
print("回数：", counter)
print("x = ", x_n_1)
print('--------------------------------')
x_n = 5
x_n_1 = 6
counter = 1
while math.fabs(x_n_1 - x_n) > 1e-6:
    x_n = x_n_1
    x_n_1 = x_n - (g(x_n))/(g_D(x_n))
#     print("x",counter," = ", x_n_1)
    counter += 1
print("回数：", counter)
print("x = ", x_n_1)

plt.figure(figsize=(10, 10))
x = np.arange(-1, 10, 0.1)
y = np.exp(-x)+x**2-6*x
plt.grid(axis="both", linestyle='-.')
plt.plot(x, y)
plt.show()
```

# 課題3

```python
def h(x):
    return x * math.exp(-x/2) -  math.exp(-x)


def h_D(x):
    return ((2 - x)/2) * math.exp(-x/2) + math.exp(-x)

x_n = 0
x_n_1 = 1
counter = 1
while math.fabs(x_n_1 - x_n) > 1e-6:
    x_n = x_n_1
    x_n_1 = x_n - (h(x_n))/(h_D(x_n))
#     print("x",counter," = ", x_n_1)
    counter += 1
print("回数：", counter)
print("x = ", x_n_1)    

plt.figure(figsize=(10, 10))
x = np.arange(-1, 100, 0.1)
y = np.exp(-x/2)- np.exp(-x)
plt.grid(axis="both", linestyle='-.')
plt.plot(x, y)
plt.show()
```
