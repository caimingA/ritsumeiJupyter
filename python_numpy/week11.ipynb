# 課題2

```python
import numpy as np
import matplotlib.pyplot as plt

sampleNum = 10000
mu = 0
sigma = 1
np.random.seed(0)
s = np.random.normal(mu, sigma, sampleNum)
width = 8 / 50
plt.figure(figsize=(10, 10))
n, bins_limits, patches = plt.hist(s, 50, density=True, facecolor='blue', edgecolor="black")
plt.plot(bins_limits[1: ] - (width / 2), n, '--', color='r')
plt.show()
```

# 課題3 

```python
import random

sampleNum = 10000
counter = 0
plt.figure(figsize=(8, 8))

x1 = list()
y1 = list()
x2 = list()
y2 = list()
for i in range(sampleNum):
    x = random.random()
    y = random.random()
    if x**2 + y**2 < 1:
        counter += 1
        x1.append(x)
        y1.append(y)
    else:
        x2.append(x)
        y2.append(y)
       
print(4*counter/sampleNum)
plt.scatter(x2, y2, color='b')
plt.scatter(x1, y1, color='r')

plt.show()
```
