```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def p(x):
    Z = 1/6
    return x*(1-x) / Z
# def p(x):
#     Z = 1/12
#     return (x**2 - x + 1/4) / Z
```

```python
k = 3
result = list()
X = np.random.uniform(0, 1, 100000)
for x in X:
    u = np.random.uniform(0, k*1)
#     print(u)
    if u <= p(x):
        result.append(x)
result = np.array(result)
# print(result)
width = 1 / 50
plt.figure(figsize=(10, 10))
n, bins_limits, patches = plt.hist(
                     result
                     , 50
                     , normed=False
                     , histtype='bar'
                     , facecolor='blue'
                     , edgecolor="black"
                     )
plt.plot(bins_limits[1:]-(width/2)
         , n
         , '--'
         , color='r'
        )
plt.savefig('hatten2.png')
plt.show()
```
