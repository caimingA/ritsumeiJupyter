```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

# 課題1

```python
A = np.array([[3, 2], [5, 2]])
w, v = la.eig(A)
print(w, "\n", v)

plt.figure(figsize=(10, 10))
plt.quiver(0, 0, v[:, 0][0], v[:, 0][1], scale=1, units='xy', width=0.1, color='r')
plt.quiver(0, 0, v[:, 1][0], v[:, 1][1], scale=1, units='xy', width=0.1, color='r')

# B = np.array([v[:, 0]*w[0], v[:, 1]*w[1]])
# print(B)
C = np.array([np.dot(A, v[:, 0]), np.dot(A, v[:, 1])])
print(C)
plt.quiver(0, 0, C[0][0], C[0][1], scale=1, units='xy', width=0.1, color='b')
plt.quiver(0, 0, C[1][0], C[1][1], scale=1, units='xy', width=0.1, color='b')

plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.grid(axis="both", linestyle='-.')
plt.show()
```

# 課題2

```python
plt.figure(figsize=(10, 10))
for i in range(0, 11):
    for j in range(0, 11):
        origin = np.array([0.1*i, 0.1*j])
        point = np.dot(A, np.array([0.1*i, 0.1*j]))
        plt.scatter(origin[0], origin[1], color='r', alpha=0.6)
        plt.scatter(point[0], point[1], color='b', alpha=1)
plt.xlim((0, 8))
plt.ylim((0, 8))
plt.show()
```

```python
plt.figure(figsize=(10, 10))
origin = list()
point = list()
for i in range(0, 11):
    for j in range(0, 11):
        origin.append(np.array([0.1*i, 0.1*j]))
        point.append(np.dot(A, np.array([0.1*i, 0.1*j])))
        
origin = np.array(origin)
point = np.array(point)
plt.scatter(origin[:, 0], origin[:, 1], color='r')
plt.scatter(point[:, 0], point[:, 1], color='b')
plt.xlim((0, 8))
plt.ylim((0, 8))
plt.show()
```
