# 課題2

```python
import numpy as np
import matplotlib.pyplot as plt

sampleNum = 10000
mu = 0
sigma = 1
np.random.seed(0)
s = np.random.normal(mu, sigma, sampleNum)
width = 8 / 50
plt.figure(figsize=(10, 10))
n, bins_limits, patches = plt.hist(s, 50, normed=True, facecolor='blue', edgecolor="black")
plt.plot(bins_limits[1: ] - (width / 2), n, '--', color='r')
plt.show()
```

# 課題3 

```python
import random

sampleNum = 10000000
counter = 0
for i in range(sampleNum):
    x = random.random()
    y = random.random()
    if x**2 + y**2 < 1:
        counter += 1
        
print(4*counter/sampleNum)
```
