```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# 課題1

```python
delta_x = 0.1
a = 0
A = 1

plt.figure(figsize=(10, 10))
x0 = np.arange(0, 1.0001, 0.0001)
y0 = np.exp(x0)
plt.plot(x0, y0, color='r')
x_T = list()
y_T = list()
x = a
y = A
for i in range(11):
    x_T.append(x)
    y_T.append(y)
    y += delta_x*y 
    x += delta_x    

plt.xlim((0, 1))
plt.grid(axis="both", linestyle='-.')
plt.plot(x0, y0, color='r')
plt.scatter(x_T, y_T, color='b')
# plt.xlabel("delta_x = 0.1", fontsize=20)
plt.title("delta_x = 0.1", fontsize=20)
plt.show()

plt.figure(figsize=(10, 10))
delta_x = 0.01
x_T = list()
y_T = list()
x = a
y = A
for i in range(101):
    x_T.append(x)
    y_T.append(y)
    y += delta_x*y 
    x += delta_x

plt.xlim((0, 1))
plt.grid(axis="both", linestyle='-.')
plt.plot(x0, y0, color='r')
plt.scatter(x_T, y_T, color='b')
plt.title("delta_x = 0.01", fontsize=20)
# plt.xlabel("delta_x = 0.01", fontsize=20)
plt.show()
```

# 課題2

```python
delta_x = 0.01
a = 0
A = 0.1
k = 2

plt.figure(figsize=(10, 10))

x_T = list()
y_T = list()
x = a
y = A
for i in range(1000):
    x_T.append(x)
    y_T.append(y)
    y += delta_x*y*(1-y)
    x += delta_x

plt.grid(axis="both", linestyle='-.')
plt.plot(x_T, y_T, color='b')
plt.show()
```

# 課題3

```python
delta_x = 0.01
a = 0
A = 0.1
k = 2

plt.figure(figsize=(10, 10))

x_T = list()
y_T = list()
x = a
y = A
for i in range(1600):
    x_T.append(x)
    y_T.append(y)
    y += delta_x*(-y+math.cos(x))
    x += delta_x

plt.grid(axis="both", linestyle='-.')
plt.plot(x_T, y_T, color='b')
plt.show()
```
